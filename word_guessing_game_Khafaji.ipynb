```python
import requests

# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

wordSet = []
fileOB = open('words.txt',"r")
lines = fileOB.read().splitlines()
for line in lines:
    wordSet.extend(line.split())
```

```python
import random
def getWord():
    index=random.randint(0,113782)
    return index
    
print("\n")

def print_und(wordIndex,inList,n):
    loopIndex=0
    suGuess=0
    sumUnd=0

    while loopIndex<len(wordSet[wordIndex]):
        if wordSet[wordIndex][loopIndex] in inList:
            print(wordSet[wordIndex][loopIndex],end=" ")
            if inList[n] in wordSet[wordIndex]:
                suGuess=1
        else:
            print("_",end=" ")
            sumUnd+=1
        loopIndex+=1
    return suGuess,sumUnd

def iniPrint(wordIndex):
    loopIndex=0
    while loopIndex<len(wordSet[wordIndex]):
        print("_",end=" ")
        loopIndex+=1
    print("")
def getInput(randWord):
    life=8
    n=0 
    inputList=[]
    print("Guess the word\n")
    iniPrint(randWord)
    while life>0:
        inputList.extend(input("\nGuess letter: "))
        print("\n",end="")
        conds= print_und(randWord,inputList,n)
        if conds[0]==0:
            life-=1
        if conds[1]==0:
            print("\nCongratulations! You won the game")
            break
        else:
            print("\nyou have", life,"guesses left!")
            if life==0:
                print("You have lost the game!")
                print("word is: ",wordSet[randWord])         
        n+=1
        
    
def main():       
    randWord=getWord()
    getInput(randWord)
    
main()
```
